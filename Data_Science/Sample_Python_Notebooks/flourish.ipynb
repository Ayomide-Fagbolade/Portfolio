```python
# prompt: read /content/sample_data/questionnaire_filled_full.csv

import pandas as pd

df = pd.read_csv('/content/sample_data/questionnaire_filled_full.csv')
df.head()
```

```python
# prompt: drop columns with missing value

# Dropping columns with any missing values
df_dropped = df.dropna(axis=1)

print("Original DataFrame shape:", df.shape)
print("DataFrame shape after dropping columns with missing values:", df_dropped.shape)
print(df_dropped.head())
```

```python
# prompt: do cronbach

!pip install pingouin
import pingouin as pg

# Assuming df_dropped contains the items for which you want to compute Cronbach's alpha
# Make sure the columns in df_dropped are numeric and represent your scale items.
# If there are non-numeric columns or columns that are not part of the scale, you need to select the relevant columns first.

# Example: Let's assume the first 5 columns of df_dropped are your scale items.
# Replace with the actual column names or indices if needed.
# Selecting only the columns that start with 'B', 'C', 'D', or 'E'
scale_items = df_dropped.filter(regex='^[BCDE]\d+')


cronbach_alpha = pg.cronbach_alpha(data=scale_items)

print("\nCronbach's alpha:")
cronbach_alpha
```

```python
# prompt:  create mask for private and public school

# Assuming 'school_type' is a column in df_dropped indicating school type
# Replace 'school_type' with the actual column name
# Replace 'Private' and 'Public' with the actual values used in your column

private_school_mask = df_dropped['SchoolType'] == 'Private'
public_school_mask = df_dropped['SchoolType'] == 'Public'

# You can then use these masks to filter the DataFrame:
private_schools_df = df_dropped[private_school_mask]
public_schools_df = df_dropped[public_school_mask]

print("\nDataFrame for Private Schools:")
print(private_schools_df.head())

print("\nDataFrame for Public Schools:")
print(public_schools_df.head())
```

```python
# prompt: for each of b1, b2......E5 find mean  and standard deviation for public , private

# Selecting the relevant columns for analysis
items_to_analyze = df_dropped.filter(regex='^[BCDE]\d+')

# Calculating mean and standard deviation for private schools
private_stats = private_schools_df[items_to_analyze.columns].agg(['mean', 'std']).T
private_stats = private_stats.rename(columns={'mean': 'Private_Mean', 'std': 'Private_StdDev'})

# Calculating mean and standard deviation for public schools
public_stats = public_schools_df[items_to_analyze.columns].agg(['mean', 'std']).T
public_stats = public_stats.rename(columns={'mean': 'Public_Mean', 'std': 'Public_StdDev'})

# Combining the statistics into a single DataFrame for comparison
comparison_stats = private_stats.join(public_stats)

print("\nMean and Standard Deviation by School Type for Items B1-E5:")
comparison_stats
```

```python
# prompt: from df_dropped find average of all B for each row and all c for each row, do corellatio and regression for public, private, combined

# Find average of all B for each row
df_dropped['Avg_B'] = df_dropped.filter(regex='^B\d+').mean(axis=1)

# Find average of all C for each row
df_dropped['Avg_C'] = df_dropped.filter(regex='^C\d+').mean(axis=1)

# Recreate private and public school dataframes after adding the new columns
private_school_mask = df_dropped['SchoolType'] == 'Private'
public_school_mask = df_dropped['SchoolType'] == 'Public'

private_schools_df = df_dropped[private_school_mask]
public_schools_df = df_dropped[public_school_mask]


# Print the first few rows to show the new columns
print("\nDataFrame with Avg_B and Avg_C:")
print(df_dropped[['Avg_B', 'Avg_C']].head())

# Correlation for combined data
combined_correlation = df_dropped[['Avg_B', 'Avg_C']].corr()
print("\nCorrelation between Avg_B and Avg_C for Combined Data:")
print(combined_correlation)

# Correlation for private schools
private_correlation = private_schools_df[['Avg_B', 'Avg_C']].corr()
print("\nCorrelation between Avg_B and Avg_C for Private Schools:")
print(private_correlation)

# Correlation for public schools
public_correlation = public_schools_df[['Avg_B', 'Avg_C']].corr()
print("\nCorrelation between Avg_B and Avg_C for Public Schools:")
print(public_correlation)

# Regression for combined data (Avg_C as dependent, Avg_B as independent)
# Need statsmodels for regression
!pip install statsmodels
import statsmodels.api as sm

X_combined = df_dropped['Avg_B']
y_combined = df_dropped['Avg_C']
X_combined = sm.add_constant(X_combined) # Add a constant for the intercept
model_combined = sm.OLS(y_combined, X_combined).fit()
print("\nRegression Results for Combined Data (Avg_C ~ Avg_B):")
print(model_combined.summary())

# Regression for private schools
X_private = private_schools_df['Avg_B']
y_private = private_schools_df['Avg_C']
X_private = sm.add_constant(X_private)
model_private = sm.OLS(y_private, X_private).fit()
print("\nRegression Results for Private Schools (Avg_C ~ Avg_B):")
print(model_private.summary())

# Regression for public schools
X_public = public_schools_df['Avg_B']
y_public = public_schools_df['Avg_C']
X_public = sm.add_constant(X_public)
model_public = sm.OLS(y_public, X_public).fit()
print("\nRegression Results for Public Schools (Avg_C ~ Avg_B):")
print(model_public.summary())
```

```python
# prompt: average and s.d for d for public , private

# Find average of all D for each row
df_dropped['Avg_D'] = df_dropped.filter(regex='^D\d+').mean(axis=1)

# Recreate private and public school dataframes after adding the new columns
private_school_mask = df_dropped['SchoolType'] == 'Private'
public_school_mask = df_dropped['SchoolType'] == 'Public'

private_schools_df = df_dropped[private_school_mask]
public_schools_df = df_dropped[public_school_mask]

# Calculating mean and standard deviation for Avg_D for private schools
private_d_stats = private_schools_df['Avg_D'].agg(['mean', 'std']).T
private_d_stats = private_d_stats.rename({'mean': 'Private_Avg_D_Mean', 'std': 'Private_Avg_D_StdDev'})

# Calculating mean and standard deviation for Avg_D for public schools
public_d_stats = public_schools_df['Avg_D'].agg(['mean', 'std']).T
public_d_stats = public_d_stats.rename({'mean': 'Public_Avg_D_Mean', 'std': 'Public_Avg_D_StdDev'})

print("\nAverage and Standard Deviation for Avg_D by School Type:")
print(private_d_stats)
public_d_stats
```

```python
# prompt: welch t_test , one tail if possible to compare c in public and private

# Perform Welch's t-test comparing 'Avg_C' between public and private schools
# A one-tailed test checks if the mean of one group is significantly greater (or less) than the mean of the other.
# Let's assume we want to test if the mean of Avg_C is *different* in private vs public schools (two-tailed).
# If you need a one-tailed test (e.g., checking if private > public or private < public), you would adjust the `alternative` parameter.
# 'two-sided' is the default. Use 'greater' or 'less' for one-tailed tests.

# Example for a two-sided test (checking for a difference):
ttest_result_c_two_sided = pg.ttest(private_schools_df['Avg_C'], public_schools_df['Avg_C'], correction=True, alternative='two-sided')

print("\nWelch's t-test comparing Avg_C between Private and Public Schools (Two-sided):")
print(ttest_result_c_two_sided)

# Example for a one-sided test (checking if Private Avg_C is greater than Public Avg_C):
ttest_result_c_greater = pg.ttest(private_schools_df['Avg_C'], public_schools_df['Avg_C'], correction=True, alternative='greater')

print("\nWelch's t-test comparing Avg_C between Private and Public Schools (One-sided: Private > Public):")
print(ttest_result_c_greater)

# Example for a one-sided test (checking if Private Avg_C is less than Public Avg_C):
ttest_result_c_less = pg.ttest(private_schools_df['Avg_C'], public_schools_df['Avg_C'], correction=True, alternative='less')

print("\nWelch's t-test comparing Avg_C between Private and Public Schools (One-sided: Private < Public):")
ttest_result_c_less
```

```python
# prompt: repeat but for B

# Perform Welch's t-test comparing 'Avg_B' between public and private schools
# Using a two-tailed test as an example:
ttest_result_b_two_sided = pg.ttest(private_schools_df['Avg_B'], public_schools_df['Avg_B'], correction=True, alternative='two-sided')

print("\nWelch's t-test comparing Avg_B between Private and Public Schools (Two-sided):")
print(ttest_result_b_two_sided)

# Example for a one-sided test (checking if Private Avg_B is greater than Public Avg_B):
ttest_result_b_greater = pg.ttest(private_schools_df['Avg_B'], public_schools_df['Avg_B'], correction=True, alternative='greater')

print("\nWelch's t-test comparing Avg_B between Private and Public Schools (One-sided: Private > Public):")
print(ttest_result_b_greater)

# Example for a one-sided test (checking if Private Avg_B is less than Public Avg_B):
ttest_result_b_less = pg.ttest(private_schools_df['Avg_B'], public_schools_df['Avg_B'], correction=True, alternative='less')

print("\nWelch's t-test comparing Avg_B between Private and Public Schools (One-sided: Private < Public):")
ttest_result_b_less
```
