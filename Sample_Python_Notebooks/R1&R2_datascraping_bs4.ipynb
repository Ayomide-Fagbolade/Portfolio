```python
import pandas as pd
from bs4 import BeautifulSoup as bs

def scrape_universities(response, table_caption,x):
    """
    Scrapes university name, website, and control type from a table in the HTML response.

    Parameters:
    - response: The HTTP response object containing the HTML content.
    - table_caption: The caption text of the table to scrape.

    Returns:
    - DataFrame: A pandas DataFrame containing the scraped data.
    """
    # Parse the content using BeautifulSoup
    soup = bs(response.text, 'html.parser')

    # Initialize an empty list to store the results
    university_data = []

    # Find all tables with the 'wikitable' class
    tables = soup.find_all('table', {'class': 'wikitable'})

    # Loop through each table to check for the desired caption
    for table in tables:
        # Find the caption inside the table
        caption = table.find('caption')

        # Check if the caption exists and matches the table_caption
        if caption and caption.get_text(strip=True) == table_caption:
            print("Table with matching caption found!")

            # Loop through each row in the table (skipping the header row)
            for row in table.find_all('tr')[1:]:
                columns = row.find_all('td')

                # Ensure the row has enough columns to avoid errors
                if len(columns) > 2:
                    university_name = columns[0].get_text(strip=True)

                    # Check if there is a website link in the first column
                    website = columns[0].find('a')['href'] if columns[0].find('a') else 'No website'

                    # Extract the control type from the last column
                    control = columns[x].get_text(strip=True)

                    # Add the data to the list
                    university_data.append({
                        'Name': university_name,
                        'Control': control,
                        'link': website
                    })

            # Break the loop once the desired table is processed
            break

    # Create a DataFrame from the list
    df = pd.DataFrame(university_data)

    return df


```

```python
# Example usage:

table_caption = 'Doctoral Universities – Very high research spending and doctorate production'
scltable = scrape_universities(response, table_caption,3)
kcltable = scrape_universities(response, "Doctoral Universities – High research activity",1)
# Display the DataFrame
print(len(scltable),len(kcltable))
scltable
```
