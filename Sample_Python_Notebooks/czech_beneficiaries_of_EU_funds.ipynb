```python
# prompt: read all csv in /content/sample_data/data  to seperate df

import glob
import pandas as pd

data_path = '/content/sample_data/data'
csv_files = glob.glob(f"{data_path}/*.csv")

dfs = {}
for csv_file in csv_files:
  df_name = csv_file.split('/')[-1].replace('.csv', '')
  dfs[df_name] = pd.read_csv(csv_file)

```

```python
translation_map = {
    'Jméno / Název': 'Name',
    'Jméno/Název': 'Name',
    'Obec': 'Municipality',
    'Okres': 'District',
    'Fond': 'Fund',
    'Opatření': 'Measure',
    'Zdroje ČR': 'CZ_Funding',
    'Zdroje EU': 'EU_Funding',
    'Celkem (Kč)': 'Total_(CZK)',
    'Typ platby': 'Payment_Type',
    'IČ': 'ID',
    'Datum nabytí právní moci': 'Date_of_Legal_Effect',
    'Typ fondu': 'Fund_Type',
    'Celkem CZK': 'Total_CZK',
}
```

```python
# prompt: use translation map to translate all column headings indfs

for df_name, df in dfs.items():
    dfs[df_name] = df.rename(columns=translation_map)

for df_name, df in dfs.items():
  print(f"--- {df_name} ---")
  print(df.head())
```

```python
# prompt: extract last 4 chars in key name of each except '2022_data' where you  extract first 4 instead ,which you insert into a new year column for each

for df_name, df in dfs.items():
    if df_name == '2022_data':
        dfs[df_name]['Year'] = df_name[:4]
    else:
        dfs[df_name]['Year'] = df_name[-4:]

for df_name, df in dfs.items():
    print(f"--- {df_name} ---")
    print(df.head())
```
