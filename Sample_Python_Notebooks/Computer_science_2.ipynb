```python
# prompt: read all the csv in /content/sample_data/app_file to different df

import pandas as pd
import os

directory_path = '/content/sample_data/app_file'
dfs = {}

# List all files in the directory
all_files = os.listdir(directory_path)

# Filter for CSV files and store them in a list
csv_files = [f for f in all_files if f.endswith('.csv')]

# Read each CSV file into a separate DataFrame and store it in the dictionary
for csv_file in csv_files:
    file_path = os.path.join(directory_path, csv_file)
    # Create a descriptive key for the dictionary (e.g., filename without extension)
    df_name = os.path.splitext(csv_file)[0]
    dfs[df_name] = pd.read_csv(file_path)

# You can now access each DataFrame by its name, e.g., dfs['california_housing_train']
print(dfs.keys()) # To see the names of the dataframes
# To see the head of a specific dataframe
```

```python
# prompt: create a pivot for allocated survey on user id with response survey

# Create a pivot table
# The index will be the 'user_id', the columns will be 'response_survey',
# and the values will be 'response_value'.
# We will use the first occurrence of 'response_value' for each user_id/response_survey combination.
pivot_table_allocation = allocated_survey.pivot_table(index='user_id',
                                           columns='response_survey',
                                           values='response_value')

pivot_table_ranking = ranked_survey.pivot_table(index='user_id',
                                           columns='response_survey',
                                           values='response_value')

print(pivot_table_ranking.head())
print(pivot_table_allocation.head())
```

```python
# prompt: for pivot_table_ranking and pivot_table_allocation add a new column for score which is average across rows

pivot_table_ranking['score'] = pivot_table_ranking.mean(axis=1)
pivot_table_allocation['score'] = pivot_table_allocation.mean(axis=1)

print("\nPivot Table Ranking with Score:")
print(pivot_table_ranking.head())

print("\nPivot Table Allocation with Score:")
print(pivot_table_allocation.head())
```

```python
# prompt: run a normality test on the score colu,n of each

from scipy import stats

print("\nNormality Test for 'score' in Pivot Table Ranking:")
# Drop any potential NaN values in the 'score' column before testing
ranking_score = pivot_table_ranking['score'].dropna()
shapiro_test_ranking = stats.shapiro(ranking_score)
print(f"Shapiro-Wilk Test (Ranking): Statistic={shapiro_test_ranking.statistic:.4f}, p-value={shapiro_test_ranking.pvalue:.4f}")

print("\nNormality Test for 'score' in Pivot Table Allocation:")
# Drop any potential NaN values in the 'score' column before testing
allocation_score = pivot_table_allocation['score'].dropna()
shapiro_test_allocation = stats.shapiro(allocation_score)
print(f"Shapiro-Wilk Test (Allocation): Statistic={shapiro_test_allocation.statistic:.4f}, p-value={shapiro_test_allocation.pvalue:.4f}")
```

```python
# prompt: run a two-tailed Welchâ€™s independent t-test to compare both and include the sd  and CI

# Perform Welch's t-test (independent samples)
# This test does not assume equal variances, which is suitable if the Shapiro-Wilk test suggests non-normality or if variance is unequal.
ttest_result = stats.ttest_ind(ranking_score, allocation_score, equal_var=False, nan_policy='omit')

print("\nWelch's Independent t-test Results:")
print(f"Test Statistic (t): {ttest_result.statistic:.4f}")
print(f"P-value: {ttest_result.pvalue:.4f}")
print(f"Degrees of Freedom: {ttest_result.df:.2f}")

# Calculate descriptive statistics including standard deviation
ranking_mean = ranking_score.mean()
ranking_std = ranking_score.std()
allocation_mean = allocation_score.mean()
allocation_std = allocation_score.std()

print(f"\nDescriptive Statistics:")
print(f"Ranking Method: Mean={ranking_mean:.4f}, Std Dev={ranking_std:.4f}")
print(f"Allocation Method: Mean={allocation_mean:.4f}, Std Dev={allocation_std:.4f}")

# Calculate Confidence Interval for the difference in means
# We will use the formula for the confidence interval of the difference between two independent means (Welch-Satterthwaite)
# from statsmodels, which is more robust for unequal variances.
!pip install statsmodels
import statsmodels.stats.api as sms

# Calculate the confidence interval for the difference in means
# The confidence interval is for (mean_ranking - mean_allocation)
cm = sms.CompareMeans(sms.DescrStatsW(ranking_score), sms.DescrStatsW(allocation_score))
ci_diff = cm.tconfint_diff(usevar='unequal') # Welch's t-test uses unequal variances

print(f"\n95% Confidence Interval for the Difference in Means (Ranking - Allocation):")
print(f"Lower Bound: {ci_diff[0]:.4f}")
print(f"Upper Bound: {ci_diff[1]:.4f}")
```
